```python
#@title Text Preprocessing in NLP

# Install necessary libraries
!pip install nltk textblob

import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from textblob import TextBlob

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('vader_lexicon')
# Download the 'punkt_tab' resource
nltk.download('punkt_tab')

```

```python
# Sample text
text = "This is a sample text with some punctuation, numbers (123), and mixed case words. Let's see how we can preprocess it!"

# 1. Tokenization
tokens = nltk.word_tokenize(text)
print("Tokens:", tokens)
```

```python
# 2. Lowercasing
lower_tokens = [token.lower() for token in tokens]
print("Lowercase Tokens:", lower_tokens)
```

```python
# 3. Removing punctuation
cleaned_tokens = [re.sub(r'[^\w\s]', '', token) for token in lower_tokens if token.isalnum()]
print("Cleaned Tokens:", cleaned_tokens)

```

```python
# 4. Stop word removal
stop_words = set(stopwords.words('english'))
print("Stop Words:", stop_words)
filtered_tokens = [token for token in cleaned_tokens if token not in stop_words]
print("Filtered Tokens (no stop words):", filtered_tokens)
```

```python
# 5. Stemming
# faster but less accuracte
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)

```

```python
# 6. Lemmatization
# Accurate but slow
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
print("Lemmatized Tokens:", lemmatized_tokens)

```

```python
# 7. Sentiment analysis example
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
sentiment_score = sia.polarity_scores("This is a very good example!")
print(f"Sentiment Score: {sentiment_score}")

textblob_sentiment = TextBlob("This is an example!").sentiment
print(f"TextBlob Sentiment: {textblob_sentiment}")
```

```python
#@title Text Classification with Scikit-learn (Colab Notebook)
!pip install scikit-learn

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Sample data
data = [
    ("This is a positive review.", "positive"),
    ("This is a negative review.", "negative"),
    ("Another positive example.", "positive"),
    ("Another negative example.", "negative"),
    ("Neutral sentence.", "neutral")
]

texts = [item[0] for item in data]
print(texts)
labels = [item[1] for item in data]
print(labels)
```

```python
# Vectorize the text data
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

X
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

X_test
```

```python
# Train a Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

```

```python
# Make predictions
y_pred = classifier.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
